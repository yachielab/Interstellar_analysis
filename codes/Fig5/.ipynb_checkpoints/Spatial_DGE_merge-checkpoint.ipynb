```r
library(Matrix)
library(Seurat)
library(dplyr)
```

```r
dirs <- list.files("/home/ha5656/work/barista/tool_test/slide_to_visium_withoutCorrection//hippo/SpaceRanger/")
# dirs <- list.files("/home/ha5656/work/barista/tool_test/slide_to_visium_ed1/hippo/SR20200611/")
dirs
```

```r
cnt <- 0
dircount <- 0
for(dir in dirs){
    if(cnt==0){
        outdir <- paste0("/home/ha5656/work/barista/tool_test/slide_to_visium_withoutCorrection//hippo/SpaceRanger/",
                         dir,
                         "/outs/filtered_feature_bc_matrix")
#         outdir <- paste0("/home/ha5656/work/barista/tool_test/slide_to_visium_ed1/hippo/SR20200611/",
#                          dir,
#                          "/outs/filtered_feature_bc_matrix")
        dge <- Read10X(outdir)
        dge <- dge[,colSums(dge)>0]
        colnames(dge) <- paste0(dir,"_",colnames(dge))
        cnt <- 1
        print(c(dir,ncol(dge)))
    }else{
        outdir <- paste0("/home/ha5656/work/barista/tool_test/slide_to_visium_withoutCorrection//hippo/SpaceRanger/",
                         dir,
                         "/outs/filtered_feature_bc_matrix")
#         outdir <- paste0("/home/ha5656/work/barista/tool_test/slide_to_visium_ed1/hippo/SR20200611/",
#                          dir,
#                          "/outs/filtered_feature_bc_matrix")
        dge.tmp0 <- Read10X(outdir)
        dge.tmp <- dge.tmp0[,colSums(dge.tmp0)>0]
        cname <- colnames(dge.tmp0)[colSums(dge.tmp0)>0]
        print(c(dir,ncol(dge.tmp)))
        if(length(cname)==1){
            print(c(dir,"true"))
            dge <- cbind(dge,dge.tmp)
            colnames(dge)[ncol(dge)] <- paste0(dir,"_",cname)
        }else{
            colnames(dge.tmp) <- paste0(dir,"_",colnames(dge.tmp))
            dge <- cbind(dge,dge.tmp)
        }
    }
    
}
```
