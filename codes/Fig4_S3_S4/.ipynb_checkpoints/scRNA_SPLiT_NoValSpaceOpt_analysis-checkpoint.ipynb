```r
library(Seurat)
library(Matrix)
library(tidyverse)

source("/home/ha5656/work/barista/paper_data/R_fig_script/src/sc_preprocess.R")

dge_base   <- readRDS("/home/ha5656/work/barista/paper_data/R_fig_script//intermediate_dataset//dge_Feb2021/vFeb2021_SPLiT_original_rawdge.rds")
dge_todrop <- readRDS("/home/ha5656/work/Interstellar_test/seurat_obj/SPLiT_to_drop_NoAlloc.drop.rds")

message("refreshing DGE")
dge_todrop        <- refresh_dge(dge_todrop,
                                 direct = T,
                                 header = F,
                                 src = 2,
                                 dest = 1,
                                 path = "/home/ha5656/work/Interstellar_test/bc_correspondence/SPLiT_to_drop_NoAlloc/split_to_drop_NoAlloc.merged.tsv")

cell.inter <- intersect(colnames(dge_base),colnames(dge_todrop))
dge_base <- dge_base[,cell.inter]
dge_todrop <- dge_todrop[,cell.inter]

dge_target_list <- list(dge_todrop)

message("determining cell set")
use_cells <- get_cells_used(dge_base,dge_target_list,percent_mito = 5,min_features=250)

dge_base   <- dge_base[,use_cells]
dge_todrop <- dge_todrop[,use_cells]

message("start seurat")
orig.intersect.seu   <- CreateSeuratObject(dge_base)
todrop.intersect.seu <- CreateSeuratObject(dge_todrop)
```
